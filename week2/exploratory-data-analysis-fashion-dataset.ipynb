# Data Cleaning and Process
* 拿掉 NAN 
* 數字處理

```python
data.columns
```

```python
# rename columns
data = data.rename(columns={'Deatils':'Details'})
```

```python
df = data.copy()
```

```python
# 拿掉 NAN 
df.dropna()
```

```python
# set string 
# size
df.Sizes = df.Sizes.str.replace('Size:','')
# discount 
df.Discount = df.Discount.str.replace('% off','')
df
```

```python
# set type
df['SellPrice'] = df['SellPrice'].astype(int)
df
```
